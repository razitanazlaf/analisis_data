## Visualization & Explanatory Analysis

### Pertanyaan 1: Pada jam berapa konsentrasi O3 tertinggi tercatat di Guanyuan?

```python
import matplotlib.pyplot as plt

# Mengambil data jam dan O3 dari dataframe
hourly_data = prsa_guanyuan_df[['hour', 'O3']]

# Menyusun daftar jam
hrs = ['00.00', '01.00', '02.00', '03.00', '04.00', '05.00', '06.00', '07.00', '08.00', 
       '09.00', '10.00', '11.00', '12.00', '13.00', '14.00', '15.00', '16.00', '17.00', 
       '18.00', '19.00', '20.00', '21.00', '22.00', '23.00']

# Membuat kamus untuk pemetaan jam
hour_dict = {i: j for i, j in enumerate(hrs)}

# Mengelompokkan data berdasarkan jam dan menghitung maksimum
hourly_data = hourly_data.groupby('hour').max().reset_index()

# Memetakan jam agar formatnya sesuai
hourly_data['hour'] = hourly_data['hour'].map(hour_dict)

# Menetapkan 'hour' sebagai indeks
hourly_data.set_index('hour', inplace=True)

# Mem-plot data
with plt.style.context('ggplot'):
    hourly_data.plot(figsize=(12, 8), legend=False, kind='barh', linewidth=.9)
```

### Pertanyaan 2: Apa bulan dengan jumlah total hujan tertinggi di Guanyuan selama periode yang dianalisis?

```python
rain_by_month = prsa_guanyuan_df.groupby('month')['RAIN'].sum()
rain_by_month.plot(kind='bar', figsize=(10,6))
plt.title("Total Curah Hujan per Bulan")
plt.show()
```

**Insight:**
- Setelah mem-plot data konsentrasi O3 berdasarkan jam, terlihat bahwa konsentrasi O3 tertinggi tercatat pada jam 17.00. Pola ini mungkin terkait dengan peningkatan aktivitas kendaraan atau sinar matahari yang mempengaruhi pembentukan ozon di atmosfer. Jam-jam tersebut biasanya menunjukkan peningkatan polusi udara akibat aktivitas manusia yang lebih intensif pada siang hari. Insight ini penting bagi otoritas lingkungan untuk merencanakan langkah mitigasi, seperti menerapkan kebijakan pengendalian emisi selama periode waktu tersebut.
- Dari visualisasi curah hujan per bulan, bulan dengan curah hujan tertinggi adalah bulan Juli. Hal ini mungkin disebabkan oleh musim hujan yang intens pada periode tersebut. Bulan dengan curah hujan tinggi juga bisa berdampak pada kualitas udara karena adanya proses pembersihan udara oleh hujan. Insight ini berguna dalam perencanaan infrastruktur dan manajemen air, terutama di daerah rawan banjir.
